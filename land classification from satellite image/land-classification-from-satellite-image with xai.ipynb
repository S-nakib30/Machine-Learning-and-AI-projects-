```python
import os
import random
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, roc_auc_score, f1_score
from sklearn.preprocessing import label_binarize
from IPython.display import display, HTML
from io import BytesIO
import base64

# Set random seeds for reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# Define dataset class
class EuroSATDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted(os.listdir(root_dir))
        self.class_to_idx = {c: i for i, c in enumerate(self.classes)}
        self.image_paths = []
        self.labels = []
        for class_name in self.classes:
            class_path = os.path.join(root_dir, class_name)
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                self.image_paths.append(img_path)
                self.labels.append(self.class_to_idx[class_name])

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image, label

# Define transformations
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

test_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load the dataset
data_dir = '/kaggle/input/eurosat-2-land-type-classification/EuroSAT_RGB'
full_dataset = EuroSATDataset(root_dir=data_dir, transform=None)
train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, test_size])

# Apply transformations after splitting
train_dataset.dataset.transform = train_transform
test_dataset.dataset.transform = test_transform

# Create data loaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

# Define models and optimizers
num_classes = len(full_dataset.classes)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

resnet18 = models.resnet18(pretrained=True)
num_ftrs = resnet18.fc.in_features
resnet18.fc = nn.Linear(num_ftrs, num_classes)
resnet18 = resnet18.to(device)
# Reduced learning rate for ResNet18
optimizer_resnet = optim.Adam(resnet18.parameters(), lr=0.0001)
criterion = nn.CrossEntropyLoss()

efficientnet = models.efficientnet_b0(pretrained=True)
num_ftrs_eff = efficientnet.classifier[1].in_features
efficientnet.classifier[1] = nn.Linear(num_ftrs_eff, num_classes)
efficientnet = efficientnet.to(device)
# Reduced learning rate for EfficientNet-b0
optimizer_eff = optim.Adam(efficientnet.parameters(), lr=0.0001)
criterion = nn.CrossEntropyLoss()

# Training function
def train_model(model, dataloader, optimizer, criterion, num_epochs, model_name):
    model.train()
    train_losses = []
    train_accuracies = []
    best_accuracy = 0.0
    best_model_state = None

    for epoch in range(num_epochs):
        epoch_loss = 0.0
        epoch_corrects = 0
        epoch_total = 0
        progress_bar = tqdm(dataloader, desc=f'Epoch {epoch+1}/{num_epochs} - {model_name}', unit='batch')
        for inputs, labels in progress_bar:
            inputs = inputs.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)

            loss.backward()
            optimizer.step()

            epoch_loss += loss.item() * inputs.size(0)
            epoch_corrects += torch.sum(preds == labels.data)
            epoch_total += labels.size(0)

            epoch_accuracy = epoch_corrects.double() / epoch_total
            progress_bar.set_postfix(loss=loss.item(), accuracy=epoch_accuracy.item())

        epoch_loss = epoch_loss / epoch_total
        epoch_accuracy = epoch_corrects.double() / epoch_total
        train_losses.append(epoch_loss)
        train_accuracies.append(epoch_accuracy.item())

        # Evaluate on the test set at the end of each epoch
        test_loss, test_accuracy, _ = evaluate_model(model, test_loader, criterion)
        print(f'{model_name} - Epoch {epoch+1} - Train Loss: {epoch_loss:.4f} - Train Acc: {epoch_accuracy:.4f} - Test Loss: {test_loss:.4f} - Test Acc: {test_accuracy:.4f}')

        if test_accuracy > best_accuracy:
            best_accuracy = test_accuracy
            best_model_state = model.state_dict()
            torch.save(best_model_state, f'best_{model_name}.pth')

    return train_losses, train_accuracies

# Evaluation function
def evaluate_model(model, dataloader, criterion):
    model.eval()
    total_loss = 0.0
    corrects = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)

            total_loss += loss.item() * inputs.size(0)
            corrects += torch.sum(preds == labels.data)
            total += labels.size(0)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / total
    accuracy = corrects.double() / total
    return avg_loss, accuracy.item(), (all_labels, all_preds)

# Function to get raw outputs (probabilities)
def get_outputs(model, dataloader):
    model.eval()
    all_outputs = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            # Apply softmax to get probabilities
            probs = torch.softmax(outputs, dim=1)
            all_outputs.extend(probs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    return np.array(all_outputs), np.array(all_labels)

# Function to plot per-class accuracy
def plot_per_class_accuracy(labels, preds, classes, title):
    class_correct = [0] * len(classes)
    class_total = [0] * len(classes)
    for i in range(len(labels)):
        label = labels[i]
        pred = preds[i]
        if label == pred:
            class_correct[label] += 1
        class_total[label] += 1

    class_accuracy = [class_correct[i] / class_total[i] if class_total[i] > 0 else 0 for i in range(len(classes))]
    plt.figure(figsize=(10, 6))
    sns.barplot(x=classes, y=class_accuracy, palette='viridis')
    plt.xlabel('Class')
    plt.ylabel('Accuracy')
    plt.title(f'{title} - Per-Class Accuracy')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Function to visualize predictions
def visualize_predictions(model, dataloader, classes, num_examples=5):
    model.eval()
    correct_examples = []
    incorrect_examples = []
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for i in range(inputs.size(0)):
                if len(correct_examples) < num_examples and preds[i] == labels[i]:
                    correct_examples.append((inputs[i].cpu(), labels[i].item(), preds[i].item()))
                elif len(incorrect_examples) < num_examples and preds[i] != labels[i]:
                    incorrect_examples.append((inputs[i].cpu(), labels[i].item(), preds[i].item()))

                if len(correct_examples) == num_examples and len(incorrect_examples) == num_examples:
                    break
            if len(correct_examples) == num_examples and len(incorrect_examples) == num_examples:
                break

    def show_examples(examples, title):
        plt.figure(figsize=(15, 5))
        for i, (img, true_label, predicted_label) in enumerate(examples):
            img = img.permute(1, 2, 0).numpy()
            mean = np.array([0.485, 0.456, 0.406])
            std = np.array([0.229, 0.224, 0.225])
            img = std * img + mean
            img = np.clip(img, 0, 1)
            plt.subplot(1, num_examples, i + 1)
            plt.imshow(img)
            plt.title(f'True: {classes[true_label]}\nPred: {classes[predicted_label]}')
            plt.axis('off')
        plt.suptitle(title)
        plt.tight_layout()
        plt.show()

    show_examples(correct_examples, f'{model.__class__.__name__} - Correct Predictions')
    show_examples(incorrect_examples, f'{model.__class__.__name__} - Incorrect Predictions')

# Train the models
num_epochs = 10
resnet_losses, resnet_accuracies = train_model(resnet18, train_loader, optimizer_resnet, criterion, num_epochs, 'ResNet18')
efficientnet_losses, efficientnet_accuracies = train_model(efficientnet, train_loader, optimizer_eff, criterion, num_epochs, 'EfficientNet-b0')

# Evaluate the best models on the test set
print("\nEvaluating best models:")
best_resnet18 = models.resnet18(pretrained=False)
best_resnet18.fc = nn.Linear(best_resnet18.fc.in_features, num_classes)
best_resnet18.load_state_dict(torch.load('best_ResNet18.pth', map_location=device))
best_resnet18 = best_resnet18.to(device)
best_resnet18.eval()
test_loss_resnet, test_accuracy_resnet, (resnet_test_labels, resnet_test_preds) = evaluate_model(best_resnet18, test_loader, criterion)
print(f'Best ResNet18 Test Accuracy: {test_accuracy_resnet:.4f}')

best_efficientnet = models.efficientnet_b0(pretrained=False)
best_efficientnet.classifier[1] = nn.Linear(best_efficientnet.classifier[1].in_features, num_classes)
best_efficientnet.load_state_dict(torch.load('best_EfficientNet-b0.pth', map_location=device))
best_efficientnet = best_efficientnet.to(device)
best_efficientnet.eval()
test_loss_eff, test_accuracy_eff, (eff_test_labels, eff_test_preds) = evaluate_model(best_efficientnet, test_loader, criterion)
print(f'Best EfficientNet-b0 Test Accuracy: {test_accuracy_eff:.4f}')

# Get outputs for ROC and AUC
resnet_outputs, resnet_labels = get_outputs(best_resnet18, test_loader)
eff_outputs, eff_labels = get_outputs(best_efficientnet, test_loader)

# Binarize labels
binarized_resnet_labels = label_binarize(resnet_labels, classes=np.arange(num_classes))
binarized_eff_labels = label_binarize(eff_labels, classes=np.arange(num_classes))

# Plotting loss and accuracy curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs + 1), resnet_losses, label='ResNet18')
plt.plot(range(1, num_epochs + 1), efficientnet_losses, label='EfficientNet-b0')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, num_epochs + 1), resnet_accuracies, label='ResNet18')
plt.plot(range(1, num_epochs + 1), efficientnet_accuracies, label='EfficientNet-b0')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# Plotting confusion matrices side by side
cm_resnet = confusion_matrix(resnet_test_labels, resnet_test_preds)
cm_eff = confusion_matrix(eff_test_labels, eff_test_preds)
classes = full_dataset.classes

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.heatmap(cm_resnet, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title('ResNet18 Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.subplot(1, 2, 2)
sns.heatmap(cm_eff, annot=True, fmt='d', cmap='Greens', xticklabels=classes, yticklabels=classes)
plt.title('EfficientNet-b0 Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.tight_layout()
plt.show()

# Plotting ROC Curves and AUC for each class (One-vs-Rest)
plt.figure(figsize=(12, 6))
for i in range(num_classes):
    fpr, tpr, _ = roc_curve(binarized_resnet_labels[:, i], resnet_outputs[:, i])
    roc_auc = roc_auc_score(binarized_resnet_labels[:, i], resnet_outputs[:, i])
    plt.plot(fpr, tpr, lw=2, label=f'ResNet18 - Class {classes[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ResNet18 - One-vs-Rest ROC Curves')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
for i in range(num_classes):
    fpr, tpr, _ = roc_curve(binarized_eff_labels[:, i], eff_outputs[:, i])
    roc_auc = roc_auc_score(binarized_eff_labels[:, i], eff_outputs[:, i])
    plt.plot(fpr, tpr, lw=2, label=f'EfficientNet-b0 - Class {classes[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('EfficientNet-b0 - One-vs-Rest ROC Curves')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# Visualize F1-Scores per class
resnet_preds = np.argmax(resnet_outputs, axis=1)
eff_preds = np.argmax(eff_outputs, axis=1)

resnet_f1_scores = f1_score(resnet_labels, resnet_preds, average=None)
eff_f1_scores = f1_score(eff_labels, eff_preds, average=None)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.barplot(x=classes, y=resnet_f1_scores, palette='Blues')
plt.title('ResNet18 - F1-Score per Class')
plt.xlabel('Class')
plt.ylabel('F1-Score')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)

plt.subplot(1, 2, 2)
sns.barplot(x=classes, y=eff_f1_scores, palette='Greens')
plt.title('EfficientNet-b0 - F1-Score per Class')
plt.xlabel('Class')
plt.ylabel('F1-Score')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)

plt.tight_layout()
plt.show()

# Visualize Correct and Incorrect Predictions for both models
visualize_predictions(best_resnet18, test_loader, full_dataset.classes)
visualize_predictions(best_efficientnet, test_loader, full_dataset.classes)
```

**my model- hybrid restnet+ CNN**

```python
class HybridResNetCNN(nn.Module):
    def __init__(self, num_classes):
        super(HybridResNetCNN, self).__init__()

        # Load pretrained ResNet18 backbone (excluding FC)
        self.resnet = models.resnet18(pretrained=True)
        self.resnet = nn.Sequential(*list(self.resnet.children())[:-2])  # Keep up to last conv layer

        # Additional custom CNN block
        self.cnn_block = nn.Sequential(
            nn.Conv2d(512, 256, kernel_size=3, padding=1),  # From ResNet output (B, 512, 7, 7)
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2),  # Output shape (B, 256, 3, 3)

            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d((1, 1))  # Output shape (B, 128, 1, 1)
        )

        # Final classification layer
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.resnet(x)           # Shape: (B, 512, 7, 7)
        x = self.cnn_block(x)        # Shape: (B, 128, 1, 1)
        x = self.classifier(x)       # Shape: (B, num_classes)
        return x
```

```python
hybrid_model = HybridResNetCNN(num_classes=num_classes).to(device)
optimizer_hybrid = optim.Adam(hybrid_model.parameters(), lr=0.0001)
hybrid_losses, hybrid_accuracies = train_model(hybrid_model, train_loader, optimizer_hybrid, criterion, num_epochs, 'HybridResNetCNN')
best_hybrid = HybridResNetCNN(num_classes=num_classes)
best_hybrid.load_state_dict(torch.load('best_HybridResNetCNN.pth', map_location=device))
best_hybrid = best_hybrid.to(device)
best_hybrid.eval()
test_loss_hybrid, test_accuracy_hybrid, (hyb_test_labels, hyb_test_preds) = evaluate_model(best_hybrid, test_loader, criterion)
print(f'Best HybridResNetCNN Test Accuracy: {test_accuracy_hybrid:.4f}')
```

```python
def get_confidence_histogram_data(model, dataloader):
    model.eval()
    all_probs = []
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)
            max_probs, preds = torch.max(probs, dim=1)

            all_probs.extend(max_probs.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    return np.array(all_probs), np.array(all_preds), np.array(all_labels)
def plot_confidence_histogram(probs, preds, labels, model_name):
    correct_mask = preds == labels
    incorrect_mask = ~correct_mask

    plt.figure(figsize=(10, 6))
    plt.hist(probs[correct_mask], bins=20, alpha=0.6, label='Correct Predictions', color='green')
    plt.hist(probs[incorrect_mask], bins=20, alpha=0.6, label='Incorrect Predictions', color='red')
    plt.xlabel('Prediction Confidence')
    plt.ylabel('Number of Samples')
    plt.title(f'{model_name} - Confidence Histogram')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
#for hybrid model :
hyb_probs, hyb_preds, hyb_labels = get_confidence_histogram_data(best_hybrid, test_loader)
plot_confidence_histogram(hyb_probs, hyb_preds, hyb_labels, 'HybridResNetCNN')

# ResNet18
res_probs, res_preds, res_labels = get_confidence_histogram_data(best_resnet18, test_loader)
plot_confidence_histogram(res_probs, res_preds, res_labels, 'ResNet18')

# EfficientNet
eff_probs, eff_preds, eff_labels = get_confidence_histogram_data(best_efficientnet, test_loader)
plot_confidence_histogram(eff_probs, eff_preds, eff_labels, 'EfficientNet-b0')
```
